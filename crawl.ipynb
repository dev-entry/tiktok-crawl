```python
def safe_prin(text, flush=False):
    try:
        if flush:
            print(text, flush=True)
        else:
            print(text)
    except:
        pass
```

```python
# translator of abstract number string to integer
def translator(number: str):
    if number.strip() == '':
        return -1
    if number[-1] == 'k' or number[-1] == 'K':
        return int(float(number[:-1].replace(",", "")) * 1000)
    elif number[-1] == 'm' or number[-1] == 'M':
        return int(float(number[:-1].replace(",", "")) * 1000000)
    else:
        return int(number.replace(",", ""))
```
